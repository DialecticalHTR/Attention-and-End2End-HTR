```python
from src.dataset.augmentation.augmentation import augment
from src.dataset.preprocessing.binarizer import Binarizer
import cv2
import matplotlib.pyplot as plt
import os
```

```python
binarizer = Binarizer()

images = [binarizer.binarize(cv2.imread(os.path.join("img", file_name))) for file_name in os.listdir("img")]
augmented_images = augment(images)

fig = plt.figure(figsize=(20, 20))
rows, columns = len(images), 2

for i in range(1, len(images)):
    fig.add_subplot(rows, columns, i * 2 - 1)
    plt.imshow(images[i])
    plt.axis('off')
    fig.add_subplot(rows, columns, i * 2)
    plt.imshow(augmented_images[i])
    plt.axis('off')
```
