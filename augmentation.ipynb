```python
from src.dataset.preprocessing.binarizer import Binarizer
import cv2
from copy import deepcopy
import matplotlib.pyplot as plt

binarizer = Binarizer(block_size=20, delta=20)
```

# Text augmentations

```python
from src.dataset.augmentation.text_augmentation import change_width, blur, erode, dilate, resize_up, resize_down


def visualize_text_augmentations(img):
    img = binarizer.binarize(img)

    fig = plt.figure(figsize=(20, 10))
    rows, columns = 8, 1

    fig.add_subplot(rows, columns, 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title("Initial image")

    fig.add_subplot(rows, columns, 2)
    plt.imshow(change_width(deepcopy(img), proportion=0.5))
    plt.axis('off')
    plt.title("Change width")

    fig.add_subplot(rows, columns, 3)
    plt.imshow(change_width(deepcopy(img), proportion=2.))
    plt.axis('off')

    fig.add_subplot(rows, columns, 4)
    plt.imshow(blur(deepcopy(img), 7))
    plt.axis('off')
    plt.title("Blur")

    fig.add_subplot(rows, columns, 5)
    plt.imshow(erode(deepcopy(img)))
    plt.axis('off')
    plt.title("Erode")

    fig.add_subplot(rows, columns, 6)
    plt.imshow(dilate(deepcopy(img)))
    plt.axis('off')
    plt.title("Dilate")

    fig.add_subplot(rows, columns, 7)
    plt.imshow(resize_up(deepcopy(img)))
    plt.axis('off')
    plt.title("Resize up")

    fig.add_subplot(rows, columns, 8)
    plt.imshow(resize_down(deepcopy(img)))
    plt.axis('off')
    plt.title("Resize down")
```

# Background augmentations

```python
from src.dataset.augmentation.background_augmentation import add_blot, fill_gradient, draw_lines, add_noise, add_stains, add_blurred_stains, add_cut_characters

def visualize_backgound_augmentations(img):
    img = binarizer.binarize(img)

    fig = plt.figure(figsize=(25, 10))
    rows, columns = 12, 1

    fig.add_subplot(rows, columns, 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title("Initial image")

    fig.add_subplot(rows, columns, 2)
    plt.imshow(add_blot(deepcopy(img), blots_num=3))
    plt.axis('off')
    plt.title("Blots")

    fig.add_subplot(rows, columns, 3)
    plt.imshow(fill_gradient(deepcopy(img), 200))
    plt.axis('off')
    plt.title("Gradient fill")

    fig.add_subplot(rows, columns, 4)
    plt.imshow(fill_gradient(deepcopy(img), 100, light=True))
    plt.axis('off')

    fig.add_subplot(rows, columns, 5)
    plt.imshow(draw_lines(deepcopy(img), 100, 2))
    plt.axis('off')
    plt.title("Draw lines")

    fig.add_subplot(rows, columns, 6)
    plt.imshow(add_noise(deepcopy(img), 100, 0, 200))
    plt.axis('off')
    plt.title("Add noise")

    fig.add_subplot(rows, columns, 7)
    plt.imshow(add_stains(deepcopy(img), 200))
    plt.axis('off')
    plt.title("Add stains")

    fig.add_subplot(rows, columns, 8)
    plt.imshow(add_stains(deepcopy(img), 100, light=True))
    plt.axis('off')

    fig.add_subplot(rows, columns, 9)
    plt.imshow(add_blurred_stains(deepcopy(img), 100))
    plt.axis('off')
    plt.title("Add blurred stains")

    fig.add_subplot(rows, columns, 10)
    plt.imshow(add_blurred_stains(deepcopy(img), 200, light=True))
    plt.axis('off')

    fig.add_subplot(rows, columns, 11)
    plt.imshow(add_cut_characters(deepcopy(img)))
    plt.axis('off')
    plt.title("Add cut characters")

    fig.add_subplot(rows, columns, 12)
    plt.imshow(add_cut_characters(deepcopy(img), under=False))
    plt.axis('off')
```

```python
visualize_text_augmentations(cv2.imread("img/augmentation.png"))
```
